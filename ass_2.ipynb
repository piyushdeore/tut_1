## Predict Month-wise Output 

### Importing necessary libraries

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

## January

```python
xtrain, xtest, ytrain, ytest = train_test_split(dataset[['YEAR']], dataset['JAN'], test_size = 0.2, random_state = 5)

regr = LinearRegression()

regr.fit(xtrain, ytrain)

ypred = regr.predict(xtest)

plt.figure(figsize = (14, 8))
plt.scatter(xtest, ytest);
plt.plot(xtest, ypred, color = 'red', linewidth = 2);
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.title('January Temperature (1901 - 2017)', fontsize = 15);
```

## Metrics for January

```python
print("R-Square Error         : ", metrics.r2_score(ytest, ypred))

print("Mean Square Error      : ", metrics.mean_squared_error(ytest, ypred))

print("Root Mean Square Error : ", np.sqrt(metrics.mean_squared_error(ytest, ypred)))

print("Mean Absolute Error    : ", metrics.mean_absolute_error(ytest, ypred))
```

## February

```python
xtrain, xtest, ytrain, ytest = train_test_split(dataset[['YEAR']], dataset['FEB'], test_size = 0.2, random_state = 5)

regr.fit(xtrain, ytrain)

ypred = regr.predict(xtest)

plt.figure(figsize = (14, 8))
plt.scatter(xtest, ytest);
plt.plot(xtest, ypred, color = 'red', linewidth = 2);
plt.ylabel('Temperature', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.title('February Temperature (1901 - 2017)', fontsize = 15);
```

## Metrics for February

```python
print("R-Square Error         : ", metrics.r2_score(ytest, ypred))

print("Mean Square Error      : ", metrics.mean_squared_error(ytest, ypred))

print("Root Mean Square Error : ", np.sqrt(metrics.mean_squared_error(ytest, ypred)))

print("Mean Absolute Error    : ", metrics.mean_absolute_error(ytest, ypred))
```

## Plotting subplots for all months 

```python
months = [] # List of months

for i in range(1, 13):
    months.append(dataset.columns[i])
```

```python
# Constructing 4 x 3 subplots for all 12 months
fig, ax = plt.subplots(4,3, figsize = (14,14))

col = 1  # column number for accessing column of every month 
num = 0  # index of months list

# Metrics for individual months
r_square = []  # List of individual r square scores
mse = []       # List of individual mean square errors
rmse = []      # List of individual root mean square errors
mae = []       # List of individual mean absolute errors

for i in range(4):
    for j in range(3):
        xtrain, xtest, ytrain, ytest = train_test_split(dataset[['YEAR']], dataset.iloc[:,col], test_size = 0.2, random_state = 5)

        regr.fit(xtrain, ytrain)

        ypred = regr.predict(xtest)
        
        # Plotting individual scatter plots
        ax[i][j].scatter(xtest, ytest);
        ax[i][j].plot(xtest, ypred, color = 'red', linewidth = 2);
        
        # Calculating metrics
        r_square.append(metrics.r2_score(ytest, ypred))

        mse.append(metrics.mean_squared_error(ytest, ypred))

        rmse.append(np.sqrt(metrics.mean_squared_error(ytest, ypred)))

        mae.append(metrics.mean_absolute_error(ytest, ypred))
        
        # Setting title of subplots
        ax[i][j].set_title(months[num], fontsize = 15)
        
        # adjusting position of subplots
        plt.subplots_adjust(bottom = 3, top = 4)
        
        # Incrementing column number and index values
        col += 1
        num += 1
        
```

```python
# Displaying metrics of all months in tabular format

print("Month\tR2\t\t\tMSE\t\t\tRMSE\t\t\tMAE")
print()

for i in range(12):
    print(months[i], end = "\t")
    print(r_square[i], end = "\t")
    print(mse[i], end = "\t")
    print(rmse[i], end = "\t")
    print(mae[i], end = "\t")
    print()
    
```
